```python
import requests
import pandas as pd
```

```python
url = r"https://jsonplaceholder.typicode.com/albums"
url2 = r"https://jsonplaceholder.typicode.com/photos"
url3 = r"https://jsonplaceholder.typicode.com/users"
```

```python
res = requests.get(url)
res2 = requests.get(url2)
res3 = requests.get(url3)
```

```python
data = res.json()
data2 = res2.json()
data3 = res3.json()
```

```python
album = pd.DataFrame(data, columns=data[0].keys())
photo = pd.DataFrame(data2, columns=data2[0].keys())
user = pd.DataFrame(data3, columns=data3[0].keys())
```

```python
result = pd.merge(user, album, on='id')
result2 = pd.merge(result, photo, on='id')
result2
```
