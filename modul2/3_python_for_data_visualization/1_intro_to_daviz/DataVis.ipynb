## **3. Lineplot**

#### Creating Lineplot Using Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns

flights = sns.load_dataset('flights')
flights.head()
```

```python
plt.figure(figsize = (8,6))
plt.plot(flights[flights['month']=='Jan']['year'],
        flights[flights['month']=='Jan']['passengers'],
         'blue'
        )
plt.title('Number of Passengers in January Yearly', size = 15)
plt.xlabel('Yearly')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
x1 = flights['year'][flights['month']=='Jan']
y1 = flights['passengers'][flights['month']=='Jan']
x2 = flights['year'][flights['month']=='Dec']
y2 = flights['passengers'][flights['month']=='Dec']

plt.style.use('seaborn-v0_8')
plt.figure(figsize=(8,6))
plt.plot(x1, y1, 'red',
        x2, y2, 'blue')
plt.title('Numbers of Passengers in January & December Yearly', size=15)
plt.xlabel('Yearly')
plt.ylabel('Number of Passengers')
plt.legend(['January Passengers', 'December Passengers'])
plt.grid(True)
plt.savefig('JanDec_Passengers_LinePlot.png')
plt.show()
```

#### Creating Lineplot Using Seaborn


```python
sns.lineplot(data=flights, x='year', y='passengers', hue='month')
plt.title('Number of Passengers Every Month', size = 20)
plt.show()
```

#### Creating Lineplot Using Pandas

```python
df_january = flights[flights['month'] == 'Jan']
df_january.plot.line(x='year', y='passengers')
plt.title('Number of Passengers in January', size = 20)
plt.show()
```

## **5. Barplot**

#### Creating Barplot Using Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
```

```python
tips_by_day = tips[['tip', 'day']].groupby(['day']).mean()
tips_by_day
```

```python
x = tips['day'].unique().tolist()
y = tips_by_day['tip']

plt.style.use('seaborn-v0_8')
plt.figure(figsize=(8,6))
plt.barh(x, y, color=['green', 'blue', 'red', 'yellow'])
plt.title('Barplot: Average Tips Everyday', size = 15)
plt.xlabel('Days')
plt.ylabel('Tip')
plt.savefig('Barplot_tipsEveryday.png')
plt.show()
```

#### Creating Barplot Using Seaborn


```python
sns.barplot(data=tips, x='day', y='tip', hue='sex')
plt.title('Barplot: Average Tips Everyday & Gender', size=15)
plt.legend(loc=0)
plt.show()
```

```python
sns.barplot(data=tips, y='day', x='tip', hue='sex')
plt.title('Horizontal Barplot: Average Tips Everyday & Gender', size=15)
plt.legend(loc=0)
plt.show()
```

```python
import numpy as np

sns.barplot(data=tips, x='tip', y='day', estimator=np.median)
plt.title('Barplot: Median Tips Everyday', size=15)
plt.show()
```

#### Creating Barplot Using Pandas

```python
tips_by_day = tips[['tip', 'day']].groupby(['day'], as_index=False).mean()
tips_by_day
```

```python
tips_by_day.plot.bar(x='day', y='tip')
plt.title('Tip by Day', size=15)
plt.show()
```

## **6. Pie Chart**

#### Creating Pie Chart Using Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
```

```python
tips_by_day = tips[['tip', 'day']].groupby(['day'], as_index=False).mean()
tips_by_day
```

```python
plt.style.use('seaborn-v0_8')
plt.figure(figsize=(8,6))
plt.pie(tips_by_day['tip'], 
        labels=tips_by_day['day'],
        autopct = '%.2f',
        explode = (0,0.2,0, 0.1)
       )
plt.axis('equal')
plt.title('Pie Chart: Average Tips Everyday', size=15)
plt.show()
```

#### Creating Pie Chart Using Pandas

```python
tips_by_day.plot.pie(y = 'tip',
                    figsize = (8,6),
                     autopct = '%.2f',
                     explode = (0, 0.2, 0, 0.1)
                    )
plt.axis('equal')
plt.title('Pie Chart: Average Tips Everyday', size=15)
plt.show()
```

## **7. Heatmap**

#### Creating Heatmap Using Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
```

```python
correlation_matrix = tips[['total_bill', 'tip', 'size']].corr()
correlation_matrix
```

```python
plt.pcolor(correlation_matrix, cmap='RdYlGn')
plt.show()
```

#### Creating Heatmap Using Seaborn

```python
# Showing Correlation Heatmap using Seaborn

plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu')
plt.show()
```

### ``Choropleth Map``

Data Source:
* Jakarta Open Data, https://data.jakarta.go.id/dataset/pelaporan-perpindahan-penduduk-keluar-provinsi-dki-jakarta-tahun-2020-berdasarkan-kelurahan/resource/134253ff-2915-44b8-a1b3-2b1869366ed0
* Jakarta by Kelurahan geo public, https://pstyd.carto.com/tables/jakarta_by_kelurahan_geo/public 


```python
df = pd.read_csv(r'D:\Purwadhika\purwadhika_repo\modul2\3_python_for_data_visualization\1_intro_to_daviz\data-jumlah-pelaporan-perpindahan-penduduk-keluar-provinsi-dki-jakarta-berdasarkan-jenis-kelamin-bulan-maret-tahun-2020.csv', sep=';')
df.head()
```

```python
df_kel = df[['kelurahan', 'jumlah']]
df_kel
# df_kel[df_kel['jumlah'] == 0.0]
```

```python
# jakarta geojson
jakarta = r'D:\Purwadhika\purwadhika_repo\modul2\3_python_for_data_visualization\1_intro_to_daviz\jakarta_by_kelurahan_geo.geojson'
jkt_location = [-6.2088, 106.8456]

jkt_map = folium.Map(location = jkt_location, zoom_start = 12)

folium.Choropleth(geo_data = jakarta,
                 data = df_kel,
                  columns = ['kelurahan', 'jumlah'],
                  key_on = 'feature.properties.name',
                  fill_color = 'YlOrRd',
                  fill_opacity = 0.7,
                  line_opacity = 0.2,
                  legend_name = 'Transmigrant Rate in Jakarta'
                 ).add_to(jkt_map)

jkt_map
```

```python
# jakarta geojson

bins = list(df_kel['jumlah'].quantile([0, 0.25, 0.5, 0.75, 1]))
five_bins = list(df_kel['jumlah'].quantile([0, 0.2, 0.4, 0.6, 0.8, 1]))
print(bins)

jakarta = r'jakarta_by_kelurahan_geo.geojson'
jkt_location = [-6.2088, 106.8456]

jkt_map = folium.Map(location = jkt_location, zoom_start = 12)

folium.Choropleth(geo_data = jakarta,
                 data = df_kel,
                  columns = ['kelurahan', 'jumlah'],
                  key_on = 'feature.properties.name',
                  fill_color = 'BuPu',
                  fill_opacity = 0.7,
                  line_opacity = 0.2,
                  legend_name = 'Transmigrant Rate in Jakarta',
                  bins = five_bins,
                  nan_fill_color = 'gray',
                  nan_fill_opacity = '0.9',
                  reset = True
                 ).add_to(jkt_map)

jkt_map
```
